```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#*.npy files are binary files to store numpy arrays
faces_image = np.load('olivetti_faces.npy') #this is a 3d numpy array
faces_target = np.load('olivetti_faces_target.npy')
```

```python
print(faces_image[:2])
print('\nShape of the faces-image matrix', faces_image.shape)
print(faces_image.shape[0])
print(faces_image.shape[1])
print(faces_image.shape[2])
```

```python
#These are the labels for the classes
faces_target #this is 1-D numpy array
```

```python

#for conversion of the 3d matrix(faces_image) to 2d matrix(faces_data) we can use np.reshape as 
#numpy.reshape(a, (a.shape[0], a.shape[1]*a.shape[2])) should be used.
#reshaping the array from 3d to 2d so that afterwards we can write the 2d array to csv file
faces_data = faces_image.reshape(faces_image.shape[0], faces_image.shape[1] * faces_image.shape[2])
print(faces_data.shape)
print(faces_data[:2])
```

```python
plt.imshow(faces_image[34],cmap='gray')
plt.show()
```

```python
n_samples = faces_image.shape[0]
print('no of samples:',n_samples)
X = faces_data
n_features = faces_data.shape[1]
print('no of features:',n_features)
#since we have 10 examples for each 40 class so total images or samples = 40 * 10 = 400
#each image is of 64 * 64 pixels = 4096 pixels
```

```python
# the label to predict is the id of the person
y = faces_target
n_classes = faces_target.shape[0]
#merged.to_csv("output.csv", index=False)
#see that the above made 2d array i.e. face_data is stored in the form of dataframe
df= pd.DataFrame(X)
df['target'] = pd.Series(y, index=df.index) #cocatenating the target as well
df

#In the below dataframe we are having 4096 columns because each image is of 64 * 64 pixels
```

```python
#Writing an object to a comma-separated values (csv) file in a specified path.
df.to_csv("face_data.csv", index=False)
```
